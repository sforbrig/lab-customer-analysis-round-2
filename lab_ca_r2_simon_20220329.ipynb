```python
mca['vehicle type'].unique()
```

```python
mca_clean = mca.drop(['vehicle type'], axis=1)
mca_clean
```

```python
import statistics as stats
stats.mode(mca['state'])
```

```python
mca['state'] = mca['state'].fillna('California')
```

```python
mca['state'] = mca['state'].fillna(stats.mode(mca['state']))
```

```python
mca['response'] = mca['response'].fillna(stats.mode(mca['response']))
mca['months since last claim'] = mca['months since last claim'].fillna(stats.mode(mca['months since last claim']))
mca['number of open complaints'] = mca['number of open complaints'].fillna(stats.mode(mca['number of open complaints']))
mca['vehicle class'] = mca['vehicle class'].fillna(stats.mode(mca['vehicle class']))
mca['vehicle size'] = mca['vehicle size'].fillna(stats.mode(mca['vehicle size']))
```

```python
mca_clean['effective to date_new'] = mca_clean['effective to date']
mca_clean.head()
```

```python
mca_clean['effective to date_new'] = pd.to_datetime(mca_clean['effective to date_new'], errors='coerce')
mca_clean.dtypes
```

```python
mca_clean['effective to date_new'] = pd.DatetimeIndex(mca_clean['effective to date_new']).month
mca_clean.head()
```
